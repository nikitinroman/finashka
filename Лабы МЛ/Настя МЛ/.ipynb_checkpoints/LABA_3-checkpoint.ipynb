### Теперь давайте приступим к моделям. Так как у нас задача регрессии, то и методы нам надо использовать соответственные. Будем использовать LinearRegression, DecisionTreeRegressor и KNeighborsRegressor

```python
#Импорт

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
import numpy as np
import timeit
```

```python
#Этот метод просто для построения графика.

def train_test_plots(clf,X_train,X_test,Y_train,Y_test,print_score = True): 
    
    training_scores = []
    test_scores = []
    numbers = []

    for i in np.linspace(20, len(X_train), 20): #Берем делаем список с помощью linespace 

        X = X_train[:int(i)] #срез
        Y = Y_train[:int(i)] #срез
        clf.fit(X, Y) #фитим

        #записываем данные после фита
        numbers.append(int(i))
        training_scores.append(clf.score(X_train, Y_train))
        test_scores.append(clf.score(X_test, Y_test))

    plt.figure(figsize=(20, 10))
    plt.plot(numbers, training_scores, label="training scores")
    plt.plot(numbers, test_scores, label="test scores")
    plt.legend(loc="best")
    
    if print_score == True:
        
        train_score=clf.score(X_train, Y_train)
        test_score=clf.score(X_test, Y_test)
        print("Точность на обучающей выборке: {0}".format(train_score))
        print("Точность на тестовой выборке: {0}".format(test_score))
        
        return (train_score, test_score)
```

```python
line = Pipeline([
    ('scaler',StandardScaler()),
    ('clf',LinearRegression())
])
   
start_time = timeit.default_timer() 
l_tr_score, l_te_score = train_test_plots(line,X_train,X_test,Y_train,Y_test,print_score=True)
l_time = f'{timeit.default_timer()-start_time}s'
```

```python
tree = Pipeline([
    ('scaler',StandardScaler()),
    ('clf',DecisionTreeRegressor())
])
    
start_time = timeit.default_timer()    
t_tr_score, t_te_score = train_test_plots(tree,X_train,X_test,Y_train,Y_test,print_score=True)
t_time = f'{timeit.default_timer()-start_time}s'
```

```python
neigh = Pipeline([
    ('scaler',StandardScaler()),
    ('clf',KNeighborsRegressor())
])
    
start_time = timeit.default_timer()    
n_tr_score, n_te_score = train_test_plots(neigh,X_train,X_test,Y_train,Y_test,print_score=True)
n_time = f'{timeit.default_timer()-start_time}s'
```

```python
total = pd.DataFrame(
    {'Name':['LinearRegression','DecisionTreeRegressor','KNeighborsRegressor'],
    'Train Score':[l_tr_score,t_tr_score,n_tr_score],
    'Test Score':[l_te_score,t_te_score,n_te_score],
    'Time':[l_time,t_time,n_time],
    'Test_size':[0.2,0.2,0.2]}
)
total
```

```python
Y=fifa1['Overall'].round(3) #Наша целевая переменная
X=fifa1.drop(['Overall'],axis=1).round(3) # Остальной датасет
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25)
```

Задаю сетку гиперпараметров для Случайного леса

```python
param_grid = {
    'max_depth': [4,5,6,8],
    'max_features': [2, 3, 4, 5, 7],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [5, 7, 9],
    'n_estimators': [100, 200, 250, 300]
}
```

Перебираю гиперпараметры на гриде для случайного леса

```python
regressor = RandomForestRegressor()
GS_rfr = GridSearchCV(estimator= regressor, param_grid= param_grid, cv= 3, n_jobs=- 1, verbose=True)
GS_rfr.fit(X_train, Y_train)
```

Вывожу результаты лучшей модели на различных метриках качества

```python
predict_val_rf = GS_rfr.predict(X_test)
print("Точность на трэйновом множестве: ", GS_rfr.score(X_train, Y_train))
print("Точность на тестовом множестве: ", GS_rfr.score(X_test, Y_test))
print("max_error : ", max_error(Y_test, predict_val_rf))
print("mean_absolute_error : ", mean_absolute_error(Y_test, predict_val_rf))
print("mean_squared_error : ", np.sqrt(mean_squared_error(Y_test, predict_val_rf)))
```

Лес деревьев перебил предыдущие результаты более, чем на 3% в score. Дальше попробую буст, так как данная модель еще лучше 

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
param_grid = {'n_estimators': [50, 100, 180, 250],
              'learning_rate': [0.1, 0.2, 0.3], 
              'max_depth': [5, 6, 8],
              'colsample_bytree': [0.6, 0.7, 1],
              'gamma': [0.07, 0.1, 0.13, 0]}

booster = XGBRegressor()
booster_grid_search = GridSearchCV(booster, param_grid, cv=3, n_jobs=-1, verbose=True)

booster_grid_search.fit(X_train, Y_train)

print(booster_grid_search.best_params_)
```

```python
predict_val_b = booster_grid_search.predict(X_test)
print("Точность на трэйновом множестве: ", booster_grid_search.score(X_train, Y_train))
print("Точность на тестовом множестве: ", booster_grid_search.score(X_test, Y_test))
print("max_error : ", max_error(Y_test, predict_val_b))
print("mean_absolute_error : ", mean_absolute_error(Y_test, predict_val_b))
print("mean_squared_error : ", np.sqrt(mean_squared_error(Y_test, predict_val_b)))
```

Далее попробую аналог обучения без учителя- ручное удаление фич при помощи feature_importances_ у буста

```python
booster = XGBRegressor(colsample_bytree= 1, gamma= 0, learning_rate= 0.1, max_depth= 6, n_estimators= 180)
booster.fit(X_train, Y_train)
predict_val_b = booster.predict(X_test)

print("Точность на трэйновом множестве: ", booster.score(X_train, Y_train))
print("Точность на тестовом множестве: ", booster.score(X_test, Y_test))
print("max_error : ", max_error(Y_test, predict_val_b))
print("mean_absolute_error : ", mean_absolute_error(Y_test, predict_val_b))
print("mean_squared_error : ", np.sqrt(mean_squared_error(Y_test, predict_val_b)))
```

Результат еще лучше, чем у случайного леса, а относительно прошлых моделей- на 7%

```python
feat_importances = pd.Series(booster.feature_importances_, index=X.columns)
feat_importances.nlargest(15).sort_values().plot(kind='barh', figsize=(10,5))
plt.xlabel('Топ 15 главных вичей с XGboost');
```

```python
for i in range(5,32):
    
    X_train, X_test, Y_train, Y_test = train_test_split(X[feat_importances.nlargest(i).sort_values().keys()],
                                                        Y, test_size=0.25)
    
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test  = sc.transform(X_test)
    
    booster = XGBRegressor(colsample_bytree= 1, gamma= 0, learning_rate= 0.1, max_depth= 6, n_estimators= 180)
    booster.fit(X_train, Y_train)
    predict_val_b = booster.predict(X_test)
    
    print(i, " -топ фичей", feat_importances.nlargest(i).sort_values().keys())
    print("Точность на трэйновом множестве: ", booster.score(X_train, Y_train))
    print("Точность на тестовом множестве: ", booster.score(X_test, Y_test))
    print("max_error : ", max_error(Y_test, predict_val_b))
    print("mean_absolute_error : ", mean_absolute_error(Y_test, predict_val_b))
    print("mean_squared_error : ", np.sqrt(mean_squared_error(Y_test, predict_val_b)))
```

```python
for i in range(1,8):
    pca = PCA(n_components=i)
    new_features = StandardScaler().fit_transform(X)
    new_features = pca.fit_transform(new_features)
    
    X_train, X_test, Y_train, Y_test = train_test_split(new_features, Y, test_size=0.25)
    
    booster = XGBRegressor(colsample_bytree= 1, gamma= 0, learning_rate= 0.1, max_depth= 6, n_estimators= 180)
    booster.fit(X_train, Y_train)
    predict_val_b = booster.predict(X_test)

    print(i, " -компонент")
    print("Точность на трэйновом множестве: ", booster.score(X_train, Y_train))
    print("Точность на тестовом множестве: ", booster.score(X_test, Y_test))
    print("max_error : ", max_error(Y_test, predict_val_b))
    print("mean_absolute_error : ", mean_absolute_error(Y_test, predict_val_b))
    print("mean_squared_error : ", np.sqrt(mean_squared_error(Y_test, predict_val_b)))
```

Вывод- лучшая модель- XGBoost, выигрывает по всем возможным метрикам качества
