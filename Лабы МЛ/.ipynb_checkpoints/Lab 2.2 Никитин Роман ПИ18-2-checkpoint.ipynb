```python
import sklearn as sk
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, KFold
from scipy.stats import sem
from sklearn.metrics import plot_precision_recall_curve
from sklearn.metrics import average_precision_score
from sklearn.metrics import precision_recall_curve
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve
```

```python
breast_cancer = load_breast_cancer()
# я не стал обрезать фичи, как мы делали это в прошлой лабе. 
X = breast_cancer.data
y = breast_cancer.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=43)
```

```python
clf = Pipeline([
    ('svc', SVC(kernel="linear")),
])
```

```python
def evaluate_cross_validation(clf, X, y, K):
    cv = KFold(K, shuffle=True, random_state=0).get_n_splits(len(y))
    scores = cross_val_score(clf, X, y, cv=cv)
    return np.mean(scores)
```

```python
train_scores = []
test_scores = []
predictions = []

def calc_params(X_train, X_test, y_train, y_test, clf, alphas, parameter, K, dataset = 1):
    best_id = 0
    best_sum = 0
    best_razn = 0
    print(clf.get_params().keys(), '\n Таким образом мы можем вывести все параметры, которые мы можем подобрать схожим образом')
    
    for i in range(len(alphas)):
        print(i)
        if dataset == 1:
            clf.set_params(svc__C=alphas[i])
        elif dataset == 2:
            clf.set_params(clf__C=alphas[i])
        #clf.set_params(**{'svc__C':0.01}) можно и так, передавать словарём- кому как больше нравится.
        #тут можно и без пайплайна, по сути, но пусть будет
        train_score = evaluate_cross_validation(clf, X_train, y_train, 3)
        train_scores.append(train_score)
        clf.fit(X_train, y_train)
        predicted = clf.predict(X_test)
        predictions.append(predicted)
        test_score = clf.score(X_test, y_test)
        test_scores.append(test_score)
        if best_sum < (test_score + train_score) and best_razn < abs(train_score - test_score):
            best_id = i
            best_sum = test_score + train_score
            best_razn = train_score - test_score            
    
    plt.figure(figsize=(15, 10))
    plt.xlabel('svc__C')
    plt.ylabel('Score')
    plt.title('Поиск лучшего параметра {parameter}')
    plt.plot(alphas, train_scores, label="training scores")
    plt.plot(alphas, test_scores, label="test scores")
    plt.legend(loc="best")
    
    return train_scores, test_scores, best_id, predictions
```

```python
alphas = np.logspace(-7, 2, 15)

train_scores, test_scores, best_id, predictions = calc_params(X_train, X_test, y_train, y_test, clf, alphas, 'svc__C', 3)
print('\n Трейновое: \n', train_scores,'\n Тестовое: \n', test_scores)
print('Лучшее:', alphas[best_id], '\n Трейновое: \n', train_scores[best_id], '\n Тестовое: \n', test_scores[best_id])
```

Оценка результата на нескольких метриках качества

```python
clf.set_params(svc__C=1.178768634793589, svc__probability=True)
clf.fit(X_train, y_train)
predicted = clf.predict(X_test)
print('accuracy_score:', accuracy_score(y_test, predicted))
print('precision_score:', precision_score(y_test, predicted))
print('recall_score:', recall_score(y_test, predicted))
print('f1_score:', f1_score(y_test, predicted))
```

Исходя из полученных результатов аккьюраси можно сделать вывод о том, что правильно классифицировано 97 процентов наблюдений, что является хорошим показателем качества модели. Что касается остальных метрик оценки качества модели- пресижн и реколл лежат практически в "верхнем правом углу", что свидетельствует о том, что у нас практически нет ошибок как первого, так и второго уровня при классификации. Ф1, как среднее гармоническое между пресижном и реколлом- подтверждает то, что модель получилась действительно гуд.

Построю график Precision-Recall по предиктам, полученным из функции выше.

```python
recall = []
precision = []
for i in predictions:
    recall.append(recall_score(y_test, i))
    precision.append(precision_score(y_test, i))
    
plt.step(recall, precision, where='post')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.94, 1.0])
plt.xlim([0.94, 1.0])
```

Забавный график получился... а теперь прост возьму ваш код из лабы

```python
y_score = clf.decision_function(X)

precision, recall, _ = precision_recall_curve(y.ravel(), y_score.ravel())
average_precision = average_precision_score(y, y_score)
 
plt.step(recall, precision, where='post')
 
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title(
    'Average precision score, micro-averaged over all classes: AP={0:0.2f}'
    .format(average_precision))
```

```python
print('f1_score:', f1_score(y_test, predicted))
```

Что касается оценки данной модели, можно сказать, что модель близка к идеалу, если не идеальная вовсе. Precision-recall уж больно хорош.

```python
plt.figure(figsize=(10, 8))
fpr, tpr, thresholds = roc_curve(y_test, clf.predict_proba(X_test)[:,1], pos_label=1)
lw = 2
plt.plot(fpr, tpr, lw=lw, label='ROC curve')
plt.plot([0, 1], [0, 1])
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
```

С ростом ФП ТП растет, на уровне ФП 0.2 ТП = 1, и так до конца. В левом верхнем углу видимая "ступенька" - лучший показатель модели.

Так как в SVC не предусмотрен n_jobs, выполню это задание с моделью RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
import time

times = []

for i in range(1,9):
    start_time = time.time()
    clf_rf = RandomForestClassifier(n_jobs=i)
    clf_rf.fit(X_train, y_train)
    times.append([i, time.time() - start_time])
    
print(times)
```

В данном случае мы видим, что быстрее всего модель обучилась на 7 ядрах. Быстрее, чем на 8. Но разница между 1 ядром и 7 по времени в два раза. На больших выборках, конечно. ситуация будет другой, но тем не менее очевидно, что использование параметра njobs существенно сокращает время обучения модели.

Повторите проведенный анализ для датасета 20newsgroup Используйте в конвейере векторизатор в в дополнении к классификатору. За счет этого, вы можете минимально модифицировать код программы. Обратите внимание, что в задачах множественной классификации обычно не строят кривые ROC и PR. В отдельных случаях их строят для каждого класса.

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
```

```python
news = fetch_20newsgroups(subset='all')
```

```python
vectorizer = CountVectorizer(token_pattern=r"\b[a-z0-9_\-\.]+[a-z][a-z0-9_\-\.]+\b")
X = vectorizer.fit_transform(news.data)

X_train, X_test, y_train, y_test = train_test_split(X, news.target, test_size=0.25)
```

```python
train_scores = []
test_scores = []
predictions = []
alphas = np.logspace(-7, 2, 15)

best_id = 0
best_sum = 0
best_razn = 0

#Пришлось менять цикл и делать без кросс валидации, а так же выводить принтами какие-то 
#состояния внутри цикла, ибо сие действие до жути медленное.

for i in range(len(alphas)):
    print('Стадия: ', i)
    clf = SVC(kernel='linear', verbose=True, C = alphas[i])
    print(i, ' стадия, после создания классификатора')
    clf.fit(X_train, y_train)
    print(i, ' стадия после фита')
    predicted = clf.predict(X_test)
    print(i, ' стадия после предикта')
    predictions.append(predicted)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    print(i, ' стадия после score')
    train_scores.append(train_score)
    test_scores.append(test_score)
    if best_sum < (test_score + train_score) and best_razn < abs(train_score - test_score):
        best_id = i
        best_sum = test_score + train_score
        best_razn = train_score - test_score            

plt.figure(figsize=(15, 10))
plt.xlabel('svc__C')
plt.ylabel('Score')
plt.title('Поиск лучшего параметра {parameter}')
plt.plot(alphas, train_scores, label="training scores")
plt.plot(alphas, test_scores, label="test scores")
plt.legend(loc="best")

print('\n Трейновое: \n', train_scores,'\n Тестовое: \n', test_scores)
print('Лучшее:', alphas[best_id], '\n Трейновое: \n', train_scores[best_id], '\n Тестовое: \n', test_scores[best_id])
```

Эта бодяга сверху фитовалась больше 4 часов на моей слабенькой ПеКарне. Тестовое на 0,86- вполне себе отличный результат, можно бахать в продакшн.
