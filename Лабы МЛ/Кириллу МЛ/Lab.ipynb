Обучение моделей классификации:

```python
Y = df['Attrition_Flag']
X = df.drop(['Attrition_Flag', 'Credit_Limit'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(X ,Y ,test_size=0.3 ,random_state=42)
```

Так же выбросил из датасета Credit_Limit, так как коррелирует с Avg_Open_To_Buy на 1.

Стандартизирую данные на вход в модель

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
classifier = RandomForestClassifier(n_estimators = 50, class_weight= 'balanced')
classifier.fit(x_train, y_train)
predict_val_rf = classifier.predict(x_test)


print("Accuracy : ", accuracy_score(y_test, predict_val_rf) *  100)
print("Recall : ", recall_score(y_test, predict_val_rf) *  100)
print("Precision : ", precision_score(y_test, predict_val_rf) *  100)
print(confusion_matrix(y_test, predict_val_rf))
print(classification_report(y_test, predict_val_rf))
```

Задаю сетку гиперпараметров для грида над случайным лесом

```python
param_grid = {
    'max_depth': [4,5,6,8],
    'max_features': [2, 3, 4, 5, 7],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [5, 7, 9],
    'n_estimators': [100, 200, 250, 300],
    'class_weight': ['None', 'balanced', 'balanced_subsample'],
}
```

Перебираю на гриде параметры и нахожу лучшие из них

```python
classifier = RandomForestClassifier()
GS_rfc = GridSearchCV(estimator=classifier, param_grid=param_grid, cv= 3, n_jobs=-1, verbose=True)
GS_rfc.fit(x_train, y_train)
```

```python
predict_val_rf = GS_rfc.predict(x_test)
```

```python
print("Accuracy : ", accuracy_score(y_test, predict_val_rf) *  100)
print("Recall : ", recall_score(y_test, predict_val_rf) *  100)
print("Precision : ", precision_score(y_test, predict_val_rf) *  100)
print(confusion_matrix(y_test, predict_val_rf))
print(classification_report(y_test, predict_val_rf))
```

В целом модель можно считать удачной, акьюраси достаточно велик, чтобы утверждать, что с задачей мы справились. Что касается precision/recall, тут все тоже вполне неплохо.

Попробую то же самое на xgboost, так как он сейчас в топе на kaggle. Возможно, результат будет лучше.

```python
param_grid = {'n_estimators': [240, 250, 270, 300],
              'learning_rate': [0.09, 0.1, 0.11, 0.13], 
              'max_depth': [6, 7, 8],
              'colsample_bytree': [0.6, 0.7],
              'gamma': [0.07, 0.1, 0.13]}

booster = XGBClassifier()
booster_grid_search = GridSearchCV(booster, param_grid, cv=3, n_jobs=-1, verbose=True)

booster_grid_search.fit(x_train, y_train)

print(booster_grid_search.best_params_)
```

```python
model = XGBClassifier(colsample_bytree= 0.6, gamma= 0.13, learning_rate= 0.13, max_depth= 6, n_estimators= 250)
model.fit(x_train, y_train)
predict_val_rf2 = model.predict(x_test)


print("Accuracy : ", accuracy_score(y_test, predict_val_rf2) *  100)
print("Recall : ", recall_score(y_test, predict_val_rf2) *  100)
print("Precision : ", precision_score(y_test, predict_val_rf2) *  100)
print(confusion_matrix(y_test, predict_val_rf2))
print(classification_report(y_test, predict_val_rf2))
```

```python
best_acc = 0
best_prec = 0
best_rec = 0
id_acc = 0
id_prec = 0
id_rec = 0

for i in range(1,70):
    
    booster = XGBClassifier(colsample_bytree= 0.7, gamma= 0.1, learning_rate= 0.1, max_depth= 6, n_estimators= 250, random_state=i)
    booster.fit(x_train, y_train)
    y_pred_test = booster.predict(x_test)
    
    accuracy = accuracy_score(y_test, y_pred_test) *  100
    recall = recall_score(y_test,y_pred_test) *  100
    precision = precision_score(y_test, y_pred_test) *  100
    
    if best_acc < accuracy:
        best_acc = accuracy
        id_acc = i
    if best_prec < precision:
        best_prec = precision
        id_prec = i
    if best_rec < recall:
        best_rec = recall
        id_rec = i
```

```python
print('best_acc: ', best_acc, 'id: ', id_acc)
print('best_prec: ', best_prec, 'id: ', id_prec)
print('best_rec: ', best_rec, 'id: ', id_rec)
```

Действительно, лучший результат показал Xgboost. Избавился от просадок в презижн и реколл, а так же немного улучшил акьюраси.
